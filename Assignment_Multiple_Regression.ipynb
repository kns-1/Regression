```python
import numpy as np
from matplotlib import pyplot as pp
```

```python
x = np.linspace(-5, 5, 30)
print("x= ",x)
```

```python
X, Y = np.meshgrid(x, x)
print("X= ",X)
print("Y= ",Y)
```

```python
np.random.seed(1)
```

```python
Z = -5 + 3*X - 0.5*Y + 8 * np.random.normal(size=X.shape)
print(Z)
```

```python
fig = pp.figure()
ax = fig.gca(projection='3d')
surf = ax.plot_surface(X, Y, Z, cmap=pp.cm.coolwarm,
                       rstride=1, cstride=1)
ax.view_init(20, -120)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
```

```python
pp.show()
```
